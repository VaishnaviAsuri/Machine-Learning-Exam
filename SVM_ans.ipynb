```python
import matplotlib.pyplot as plt

# Define the points and labels
points = [[3, 0], [1, 2], [0, 1]]
labels = [1, 1, -1]  # You can choose any labels as long as they represent both classes

# Plot the Cartesian coordinate system
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot the points
for i, point in enumerate(points):
    label = '+' if labels[i] == 1 else '-'
    plt.scatter(point[0], point[1], label=f'Point {i + 1} ({label}1)')

# Add labels to points
for i, txt in enumerate(labels):
    plt.annotate(f'Label: {txt}', (points[i][0], points[i][1]), textcoords="offset points", xytext=(0, 10), ha='center')

# Set plot properties
plt.title('Cartesian Coordinate System with Labeled Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given points
points = np.array([[3, 0], [1, 2], [0, 1]])

# Given weighted vector and offset
weights = np.array([0.46, 0.042])
offset =  -0.38

# Function to calculate the y values on the separation line
def separation_line(x):
    return (-weights[0] * x - offset) / weights[1]

# Plotting the points
plt.scatter(points[:, 0], points[:, 1], c='blue', label='Points')

# Generating x values for the separation line
x_values = np.linspace(min(points[:, 0]) - 1, max(points[:, 0]) + 1, 100)

# Plotting the separation line
plt.plot(x_values, separation_line(x_values), c='red', label='Separation Line')

# Adding labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```
